```python
# Create text documents for RAG

documents = []

for _, row in df.iterrows():
    text = (
        f"District: {row['district']}, "
        f"State: {row['State']}, "
        f"Office Type: {row['officetype']}, "
        f"District Delivery Rate: {row['district_delivery_rate']}"
    )

    documents.append(text)

print(documents[:5])
```

```python
# Convert documents into vectors (embeddings)  for 5k rows
from sentence_transformers import SentenceTransformer

# Use only first 5000 documents
sample_docs = documents[:5000]

# Load embedding model
model = SentenceTransformer('all-MiniLM-L6-v2')

# Create embeddings
embeddings = model.encode(
    sample_docs,
    batch_size=32,
    show_progress_bar=True
)

print("Number of vectors:", len(embeddings))
print("Dimensions of one vector:", len(embeddings[0]))
```

```python
## Create the Vector Database

import faiss  # FAISS Search for vectors
import numpy as np

# Convert embeddings to float32
embedding_array = np.array(embeddings).astype("float32")

# Get vector size
dimension = embedding_array.shape[1]

# Create FAISS index
index = faiss.IndexFlatL2(dimension)

# Add vectors
index.add(embedding_array)

print("Total vectors stored:", index.ntotal)
```

Dataset → Documents
Documents → Embeddings
Embeddings → FAISS Vector Database
Vector Database Ready for Search

```python
query = "Which districts have poor delivery performance?"

# Convert question into embedding
query_embedding = model.encode([query]).astype("float32")

# Search top 5 similar records
distances, indices = index.search(query_embedding, k=5)

print("Top matching records:\n")

for idx in indices[0]:
    print(sample_docs[idx])
```

```python
## Let's complete the "Generation" part using an LLM
context = "\n".join(
    [sample_docs[idx] for idx in indices[0]]
)

prompt = f"""
Context:
{context}

Question:
{query}

Answer:
"""
print(prompt)
```

```python
def ask_rag(question, k=5):
    # Convert question to embedding
    query_embedding = model.encode([question]).astype("float32")

    # Search similar records
    distances, indices = index.search(query_embedding, k=k)

    print(f"\nQuestion: {question}\n")
    print("Retrieved Context:\n")

    for idx in indices[0]:
        print(sample_docs[idx])
        print("-" * 80)
```

```python
ask_rag("Which districts have poor delivery performance?")
```

```python
## Let's connect your retrieval system to a simple local "LLM-style" response so it feels more like chatting with your data.
```

```python
def postal_chatbot(question, k=5):
    # Convert question to embedding
    query_embedding = model.encode([question]).astype("float32")

    # Retrieve relevant records
    distances, indices = index.search(query_embedding, k=k)

    # Build context
    context = [sample_docs[idx] for idx in indices[0]]

    # Generate answer
    print("\n" + "="*60)
    print("QUESTION:")
    print(question)

    print("\nANSWER:")
    print("Based on the postal delivery records, I found the following relevant information:\n")

    for record in context:
        print("•", record)

    print("\n" + "="*60)
```

```python
postal_chatbot("Show postal records from Telangana")
```

```python
from transformers import pipeline

generator = pipeline(
    "text-generation",
    model="Qwen/Qwen2-0.5B-Instruct"
)
```

```python
def ask_llm_rag(question, k=5):
    # Convert question to vector
    query_embedding = model.encode([question]).astype("float32")

    # Search similar records
    distances, indices = index.search(query_embedding, k=k)

    # Build context
    context = "\n".join([sample_docs[idx] for idx in indices[0]])

    # Create prompt
    prompt = f"""
You are a postal delivery analyst.

Context:
{context}

Question:
{question}

Provide a short answer based only on the context above.
"""

    # Generate answer
    response = generator(
        prompt,
        max_new_tokens=100,
        do_sample=False
    )

    print("\n===== ANSWER =====\n")
    print(response[0]["generated_text"])
```

```python
ask_llm_rag("Which districts have excellent delivery performance?")
```
